## 2739. N을 입력받은 뒤, 구구단 N단을 출력하는 프로그램을 작성하시오.

```
#공
N = int(input())
for i in range(1,10):
    print(N,"*",i,"=",N*i)
```

## 10950. 두 정수 A와 B를 입력받은 다음, A+B를 출력하는 프로그램을 작성하시오.
```
#공
import sys
T = int(input())
for i in range(T):
    A, B = map(int, sys.stdin.readline().split())
    print(A+B)
```

## 2439. 첫째 줄에는 별 1개, 둘째 줄에는 별 2개, N번째 줄에는 별 N개를 찍는 문제
```
#공
n = int(input())

for i in range(n):
    print(" "*(n-i-1),"*"*(i+1),sep="")
```
### sep
- 구분자를 설정해주는 것
- 기본 값은 빈칸

## 10871. 정수 N개로 이루어진 수열 A와 정수 X가 주어진다. 이때, A에서 X보다 작은 수를 모두 출력하는 프로그램을 작성하시오.

```
#공
N, X = map(int, input().split())
a = list(map(int, input().split()))
b=[]
for i in range(N):
    if a[i] < X:
        b.append(a[i])
print(*b)
```

### * list
- * symbol is use to print the list elements in a single line with space.
- print(*list, sep=' ')
