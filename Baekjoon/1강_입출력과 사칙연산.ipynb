
## 1000. 두 정수 A와 B를 입력받은 다음, A+B를 출력하는 프로그램을 작성하시오.

```
#공
digits = input().split()
add_result = 0
for i in digits:
    add_result = add_result + int(i)
print(add_result)
```

- `split()` method는 string을 list로 쪼갭니다.
- 사용법: `string.split(separator, maxsplit)`
- separator: 선택. string을 찢을 때 사용하는 구분자를 의미합니다. 구분자는 list에 들어가지 않습니다. 기본 값은 **띄어쓰기** 입니다.
- maxsplit: 선택. 얼마나 많이 쪼갤지를 의미합니다. 기본 값은 전체를 의미하는 **-1** 입니다.

## 1001. 두 정수 A와 B를 입력받은 다음, A-B를 출력하는 프로그램을 작성하시오.

```
#공
A,B = list(map(int, input().split()))
print(A-B)
```
### `map()`
- `map()` function은 특정 함수를 iterable 안의 각각의 객체에 적용시킵니다.
- 사용법: map(function, iterables)
- function: 필수. 적용할 함수입니다.
- iterable: 필수. sequence, collection, iterator 객체입니다.
