
## 9498. 시험 점수를 입력받아 90 ~ 100점은 A, 80 ~ 89점은 B, 70 ~ 79점은 C, 60 ~ 69점은 D, 나머지 점수는 F를 출력하는 프로그램을 작성하시오.

```
#공
score = int(input())

if score >= 90:
    print("A")
elif score >= 80:
    print("B")
elif score >= 70:
    print("C")
elif score >= 60:
    print("D")
else:
    print("F")
```

## 2753. 연도가 주어졌을 때, 윤년이면 1, 아니면 0을 출력하는 프로그램을 작성하시오.
윤년은 연도가 4의 배수이면서, 100의 배수가 아닐 때 또는 400의 배수일 때이다.

```
#공
year = int(input())

if (year%4==0 and year%100!=0) or (year%400==0):
    print(1)
else:
    print(0)
```

## 14681. 점의 좌표를 입력받아 그 점이 어느 사분면에 속하는지 알아내는 프로그램을 작성하시오.
단, x좌표와 y좌표는 모두 양수나 음수라고 가정한다.

```
#공
x_value = int(input())
y_value = int(input())

if x_value> 0 and y_value > 0:
    print(1)
elif x_value <0 and y_value > 0 :
    print(2)
elif x_value <0 and y_value < 0 :
    print(3)
else:
    print(4)
```

## 2884. 원래 설정되어 있던 알람을 45분 앞서는 시간으로 바꾸기

```
#공
H, M = list(map(int, input().split()))

if M>=45:
    print(H,M-45)
elif M<45 and H!=0:
    print(H-1,M+15)
else:
    print(H+23,M+15)
```
